```python
import pandas as pd
pd.set_option('display.max_columns', None)
url = 'https://www.sharkattackfile.net/spreadsheets/GSAF5.xls'
shark_df = pd.read_excel(url)
```

```python
shark_df.head(3)
```

```python
#Data Description

#Date: when the attack happened
#Year: the year the attack happened
#Type: the type of shark attack, if it was provoked or unprovoked
#Country: country the attack happened
#State: country state the attack happened
#Location: location that occured the attack
#Activity: In which activity the attack happened
#Name: name of the person that suffered the attack
#Sex: gender of the person that suffered the attack
#Age: age of the person that suffered the attack
#Injury: the injury of the attack
#Time: the time that the attack happen
#Species: specie of shark that attacked
#Source: where this information came from
#pdf: PDF file of the description of the attack

#href formula:
#href:
#Case Number
#Case Number.1:
#Original Order:
#Unnamed: 11:
#Unnamed: 21:
#Unnamed: 22:
```

```python
shark_df = shark_df.drop(columns = ['pdf', 'Unnamed: 11', 'href formula', 'href', 'Case Number', 'Case Number.1', 'original order', 'Source', 'Unnamed: 21', 'Unnamed: 22'])
```

```python
gender_mapping = {
    'F': 'F',
    'lli': 'F',
    'M': 'M',
    'N': 'M',
    ' M': 'M',
    'M x 2': 'M',
    '.': 'F',
    'M ': 'M'
}
shark_df['Sex'] = shark_df['Sex'].replace(gender_mapping)
```
